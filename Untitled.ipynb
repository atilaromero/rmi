```python
import serial
import time
```

```python
import math
```

```python
def flush(ser):
    time.sleep(0.1)
    msg = ser.read_all()
    print(msg)
```

```python
def testmode(ser):
    ser.write(b'testmode on\n')
    flush(ser)
```

```python
class Location:
    def __init__(self, x=0, y=0, theta=0):
        self.x = x
        self.y = y
        self.theta =theta
    def add(self, x, y , theta):
        self.x += x
        self.y += y
        self.theta += theta
        self.theta = self.theta %(math.pi*2)
    def wheels(self, l,r):
        dtheta = (r-l)/235
        dx=((r+l)/2) * math.cos(self.theta)
        dy=((r+l)/2) * math.sin(self.theta)
        self.add(dx,dy,dtheta)
    def __repr__(self):
        return '%.2f %.2f %.2f'%(self.x, self.y, self.theta/math.pi*180)
```

```python
location=Location()
location
```

```python
location.wheels(100,100+math.pi/2*235)
location
```

```python
def setmotor(ser, l, r, speed):
    ser.write(b'setmotor lwheeldist %i rwheeldist %i speed %i\n'%(l, r, speed))
    location.wheels(l,r)
    time.sleep(max(abs(l),abs(r))/speed+0.1)
    flush(ser)
def forward(ser, dist, speed):
    setmotor(ser, dist, dist, speed)
def turn(angle, speed):
    num = angle/90*186/180*math.pi
    setmotor(ser, -num, num, speed)
```

```python
def setldsrotation(ser):
    ser.write(b'setldsrotation on\n')
    flush(ser)
```

```python
def getldsscan(ser):
    flush(ser)
    ser.write(b'getldsscan\n')
    time.sleep(0.5)
    msg = ser.read_all()
    return msg.split()
```
